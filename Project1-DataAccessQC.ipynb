# **5. Trimming with fastp or trimmmomatic**
### *Total Questions: 3*
### *Total Points: 10+10+10 = 30*


---

<div class="alert alert-block alert-warning">
    <h3>5.1 What was the tools and command you used?</h3>

```python
# Download fastp
!wget http://opengene.org/fastp/fastp -O fastp

# Make fastp executable
!chmod +x fastp

#We used fastp
```

```python
# Running fastp for paired-end trimming with quality control
!./fastp -i SRR081224_1.filt.fastq.gz -I SRR081224_2.filt.fastq.gz -o SRR081224_1_trimmed.fastq.gz -O SRR081224_2_trimmed.fastq.gz --cut_front --cut_tail --cut_front_window_size 4 --cut_tail_window_size 4 --cut_front_mean_quality 20 --cut_tail_mean_quality 20 --trim_front1=15 --trim_front2=15
```

<div class="alert alert-block alert-warning">
    <h3>5.2 How did you perform the trimming?</h3>
    <p>Did you use dynamic built in features or hard coded parameters?</p>
    <p>Did you have to go back and repeat the trimming after run(s) for <b>postQC?</b></p>
</div>

### Trimming parameters

We performed the trimming using **fastp**. Our trimming process involved using both dynamic features (`--cut_front` and `--cut_tail`) and hard-coded parameters (`--trim_front1=15` and `--trim_front2=15`) to remove low-quality sequences and adapter contamination. Specifically, we trimmed low-quality bases from both the front and tail of the reads and removed the first 15 bases from each read due to the low-quality and base composition bias detected in the initial quality check using FastQC.

### Did you have to go back and repeat the trimming after run(s) for postQC?

After performing the trimming, we ran a **post-trimming quality check** using FastQC to assess the effectiveness of the trimming process. The results indicated that the quality of the reads had significantly improved, with a higher percentage of bases meeting the Q20 and Q30 thresholds. Additionally, adapter contamination was successfully removed, and the duplication rate remained low. This confirmed that the trimming was successful in addressing the quality issues identified in the initial analysis.

Since the post-trimming quality metrics were satisfactory, **we did not need to repeat the trimming**. The FastQC report showed no significant remaining issues with base quality, adapter contamination, or sequencing errors.
